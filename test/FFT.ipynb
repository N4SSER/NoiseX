```python
from scipy.fft import fft, ifft
import numpy as np
import matplotlib.pyplot as plt
import math
ts = 0.01 #Ts
t=np.arange(0,50,ts,dtype=float);
f_clean = np.sin(np.pi/3*t) + 2*np.sin(np.pi/4*t) #Target Signal
f = 20*np.random.rand(len(f_clean)) + f_clean; #Noisy signal
```

```python
n = len(t);
fhat = fft(f)
freq = 1/(ts*n)*np.arange(0,n,1)
psd = np.abs(fhat)
indices = np.where(psd>1500,1,0)
fhat = fhat*indices
denoised = ifft(fhat)
```

```python
figure, axis = plt.subplots(2, 2)
  
# For Sine Function
axis[0, 0].plot(f_clean)
axis[0, 0].set_title("Target Signal",fontsize = 7)
  
# For Cosine Function
axis[0, 1].plot(f)
axis[0, 1].set_title("Noisy input signal",fontsize = 7)
  
# For Tangent Function
axis[1, 0].plot(freq, psd)
axis[1, 0].set_title("FFT",fontsize = 7)
  
# For Tanh Function
axis[1, 1].plot(denoised)
axis[1, 1].set_title("Denoised signal",fontsize = 7)
```
